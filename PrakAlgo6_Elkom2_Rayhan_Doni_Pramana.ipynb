```python
print('@@@@ @@@@ @   @ @  @ @@@@ @@    @')
print('@  @ @  @  @ @  @  @ @  @ @ @   @')
print('@@@@ @@@@   @   @@@@ @@@@ @  @  @')
print('@ @  @  @   @   @  @ @  @ @   @ @')
print('@  @ @  @   @   @  @ @  @ @    @@')
import math

def hitung_luas_kubus(sisi):
    return 6 * sisi**2

def hitung_luas_balok(panjang, lebar, tinggi):
    return 2 * (panjang * lebar + panjang * tinggi + lebar * tinggi)

def hitung_luas_tabung(jari_jari, tinggi):
    return 2 * math.pi * jari_jari * (jari_jari + tinggi)

def hitung_luas_kerucut(jari_jari, tinggi):
    s = math.sqrt(jari_jari**2 + tinggi**2)
    return math.pi * jari_jari * (jari_jari + s)

def hitung_luas_bola(jari_jari):
    return 4 * math.pi * jari_jari**2

while True:
    print("KALKULATOR MENCARI LUAS")
    print("1. Kubus")
    print("2. Balok")
    print("3. Tabung")
    print("4. Kerucut")
    print("5. Bola")
    print("6. Keluar")

    pilihan = input("Mau yang mana: ")

    if pilihan == "6":
        print("Terima kasih!")
        break

    if pilihan in ["1", "2", "3", "4", "5"]:
        if pilihan == "1":
            sisi = float(input("Masukkan sisi : "))
            luas = hitung_luas_kubus(sisi)
            print(f"sisi = {sisi}")
        elif pilihan == "2":
            panjang = float(input("Masukkan panjang : "))
            lebar = float(input("Masukkan lebar : "))
            tinggi = float(input("Masukkan tinggi : "))
            luas = hitung_luas_balok(panjang, lebar, tinggi)
            print(f"panjang = {panjang}")
            print(f"lebar = {lebar}")
            print(f"tinggi = {tinggi}")
        elif pilihan == "3":
            jari_jari = float(input("Masukkan jari-jari : "))
            tinggi = float(input("Masukkan tinggi : "))
            luas = hitung_luas_tabung(jari_jari, tinggi)
            print(f"jari-jari = {jari_jari}")
            print(f"tinggi = {tinggi}")
        elif pilihan == "4":
            jari_jari = float(input("Masukkan jari-jari : "))
            tinggi = float(input("Masukkan tinggi : "))
            luas = hitung_luas_kerucut(jari_jari, tinggi)
            print(f"jari-jari = {jari_jari}")
            print(f"tinggi = {tinggi}")
        elif pilihan == "5":
            jari_jari = float(input("Masukkan jari-jari : "))
            luas = hitung_luas_bola(jari_jari)
            print(f"jari-jari = {jari_jari}")

        print(f"Luas = {luas:.2f}")
    else:
        print("Pilihan tidak valid. Silakan pilih kembali.")
```
